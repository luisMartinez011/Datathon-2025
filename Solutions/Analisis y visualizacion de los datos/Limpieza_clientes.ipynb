```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
clientes = pd.read_csv("base_clientes_final.csv")
clientes.head()
```

```python
# Convertir la columna a tipo datetime, sin importar el formato original
clientes["fecha_nacimiento"] = pd.to_datetime(clientes["fecha_nacimiento"], errors='coerce')

# Convertir todas las fechas al formato día/mes/año (DD/MM/YYYY)
clientes["fecha_nacimiento"] = clientes["fecha_nacimiento"].dt.strftime("%d/%m/%Y")

# Ver los primeros registros para confirmar
print(clientes[["fecha_nacimiento"]].head())
```

```python
from datetime import datetime
#Edad
clientes["fecha_nacimiento_original"] = pd.to_datetime(clientes["fecha_nacimiento"], errors='coerce')  # Convertir pero conservar la columna original
clientes["fecha_nacimiento"] = clientes["fecha_nacimiento_original"].dt.strftime("%d/%m/%Y")  # Formato correcto DD/MM/YYYY

# Calcular edad en una nueva columna
clientes["edad"] = (datetime.now() - clientes["fecha_nacimiento_original"]).dt.days // 365  # Calcular edad

# Mostrar los primeros registros para verificar
print(clientes[["fecha_nacimiento_original", "fecha_nacimiento", "edad"]].head())
```

```python
clientes["edad"].value_counts()
```

```python
clientes = clientes.drop("edad", axis=1)
```

```python
# Convertir la columna a tipo datetime, sin importar el formato original
clientes["fecha_alta"] = pd.to_datetime(clientes["fecha_alta"], errors='coerce')

# Convertir todas las fechas al formato día/mes/año (DD/MM/YYYY)
clientes["fecha_alta"] = clientes["fecha_alta"].dt.strftime("%d/%m/%Y")

print(clientes[["fecha_alta"]].head())
```

```python
# Asegurar conversión sin perder el formato original
clientes["fecha_alta_original"] = pd.to_datetime(clientes["fecha_alta"], errors='coerce')  # Mantener datetime para cálculos
clientes["fecha_alta"] = clientes["fecha_alta_original"].dt.strftime("%d/%m/%Y")  # Formato visual DD/MM/YYYY

# Calcular el tiempo de apertura
clientes["tiempo_apertura"] = (datetime.now() - clientes["fecha_alta_original"]).dt.days // 365  # Calcular años

# Mostrar resultados
print(clientes[["fecha_alta_original", "fecha_alta", "tiempo_apertura"]].head())
```

```python

clientes["fecha_nacimiento_dt"] = pd.to_datetime(clientes["fecha_nacimiento"], errors='coerce')
clientes["edad"] = (datetime.now() - clientes["fecha_nacimiento_dt"]).dt.days // 365

# Mostrar los primeros registros para verificar
print(clientes[["fecha_nacimiento", "edad"]].head())
```

```python
clientes = clientes[clientes["edad"] != 127]
```

```python
clientes["id_estado"] = clientes["id_estado"]-46
clientes["id_estado"].value_counts()
```
