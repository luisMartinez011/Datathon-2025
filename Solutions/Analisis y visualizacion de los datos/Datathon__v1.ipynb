```python

#Gráfica del sexo de la población.
genero= clientes["genero"].value_counts()

grupos = genero.index
valor = genero.values
colorsex = ["#FE6DA2", "#6D89FE"]

fig, ax = plt.subplots()
ax.pie(valor, labels = grupos, colors = colorsex, autopct = '%1.1f%%')
ax.set(title='Distribución de sexos en la población')
plt.show()
```

```python
from datetime import datetime

clientes["fecha_nacimiento"] = pd.to_datetime(clientes["fecha_nacimiento"])  # Convertir a formato datetime
clientes["edad"] = (datetime.now() - clientes["fecha_nacimiento"]).dt.days // 365  # Calcular edad

# Ver los primeros registros con la nueva columna
print(clientes[["fecha_nacimiento", "edad"]].head())


```

```python
import matplotlib.pyplot as plt

# Gráfica de la distribución de edades en la población
edad = clientes["edad"].value_counts()

grupos_1 = edad.index
valor_1 = edad.values
colorsex = ["#FE6DA2", "#6D89FE"]

fig, ax = plt.subplots()
ax.barh(grupos_1, valor_1, color=colorsex)  # Eliminé parámetros incorrectos y agregué 'color'
ax.set_xlabel("Cantidad de personas")
ax.set_ylabel("Edad")
ax.set_title("Distribución de edad en la población")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# 1. Conteo de personas por edad y genero
# ------------------------------------------------------------------
age_gender_counts = (clientes
                     .groupby(['edad', 'genero'])
                     .size()
                     .unstack(fill_value=0)
                     .sort_index())

# Asegurar que existan ambas columnas (por si acaso solo hay 'F' o 'M')
for col in ['M', 'F']:
    if col not in age_gender_counts.columns:
        age_gender_counts[col] = 0

# Series para graficar
hombres = -age_gender_counts['M']  # negativo para izquierda
mujeres =  age_gender_counts['F']  # positivo derecha
edades  =  age_gender_counts.index

# ------------------------------------------------------------------
# 2. Gráfico
# ------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(8, 10))

ax.barh(edades, hombres, color="#6D89FE", edgecolor="black", label="Hombres")
ax.barh(edades, mujeres, color="#FE6DA2", edgecolor="black", label="Mujeres")

ax.set_xlabel("Número de individuos")
ax.set_ylabel("Edad")
ax.set_title("Pirámide de población por género y edad (sin agrupar)")
ax.legend(loc="upper right")

# Eje X simétrico
max_count = max(mujeres.max(), (-hombres).max())
ax.set_xlim(-max_count*1.1, max_count*1.1)

# Opcional: ticks cada 5 años
ax.set_yticks(range(edades.min(), edades.max()+1, 5))

plt.tight_layout()
plt.show()
```

```python
# Obtener la frecuencia de cada actividad empresarial
actividad_counts = clientes["actividad_empresarial"].value_counts()

# Gráfica de barras
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(actividad_counts.index, actividad_counts.values, color="#6D89FE")

# Personalización
ax.set_xticklabels(actividad_counts.index, rotation=90)  # Rotación para que las etiquetas sean legibles
ax.set_xlabel("Actividad Empresarial")
ax.set_ylabel("Frecuencia")
ax.set_title("Distribución de Actividad Empresarial en la Población")

plt.show()
```
