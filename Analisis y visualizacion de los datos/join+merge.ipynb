```python
df = pd.read_csv("db_join (1).csv")
df.head()
```

```python
df.info()
```

```python
from datetime import datetime
# Fecha de corte
fecha_corte = pd.Timestamp('2023-01-31')
# Asegúrate de que 'fecha_nacimiento' esté en datetime
df['fecha_nacimiento'] = pd.to_datetime(df['fecha_nacimiento'], errors='coerce', dayfirst=True)
# Calcula la edad al 31 de enero de 2023
df['edad'] = df['fecha_nacimiento'].apply(
    lambda fn: fecha_corte.year - fn.year - ((fecha_corte.month, fecha_corte.day) < (fn.month, fn.day))
)
df.head()
```

```python
df.info()
```

```python
# Paso 1: Asegúrate de que 'fecha' es datetime y comercio está codificado
df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce', dayfirst=True)
df['comercio_le'] = pd.factorize(df['comercio'])[0].astype('int32')

# Paso 2: Crear DataFrame con nuevas columnas para análisis
filas = []

# Agrupa por cliente y comercio
for (cli, com), g in df.groupby(['id', 'comercio_le']):
    g = g.sort_values('fecha').reset_index(drop=True)
    g['days_diff'] = g['fecha'].diff().dt.days

    for i in range(1, len(g) - 1):
        curr = g.loc[i]
        prev = g.loc[:i-1]
        nxt  = g.loc[i+1]
        delta = (nxt['fecha'] - curr['fecha']).days

        filas.append({
            'id'                    : curr['id'],
            'fecha'                : curr['fecha'],
            'comercio_le'          : curr['comercio_le'],
            'días_desde_última'    : curr['days_diff'] or 0,
            'monto_actual'         : curr['monto'],
            'promedio_montos_previos': prev['monto'].mean() or curr['monto'],
            'periodicidad_media'   : prev['days_diff'].dropna().mean() or 0,
            'trans_previas'        : i
        })

df_features = pd.DataFrame(filas)
print("Filas generadas con features:", len(df_features))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ────────────────────── 1. UNIR df_features A df ORIGINAL ──────────────────────
df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce', dayfirst=True)
df['comercio_le'] = pd.factorize(df['comercio'])[0].astype('int32')

# Merge por ID, fecha y comercio_le
df_merged = df.merge(df_features, on=['id', 'fecha', 'comercio_le'], how='left')

# ────────────────────── 2. FILTRAR SOLO FILAS COMPLETAS PARA CORR ──────────────
# Nos quedamos solo con las filas que tienen todas las features numéricas disponibles
vars_corr = ['edad', 'genero', 'tiempo_apertura', 'monto',
             'días_desde_última', 'monto_actual',
             'promedio_montos_previos', 'periodicidad_media', 'trans_previas']

df_corr = df_merged.dropna(subset=vars_corr)

print("Filas listas para correlación:", len(df_corr))

# ────────────────────── 3. CALCULAR Y GRAFICAR CORRELACIÓN SPEARMAN ────────────
# Correlación Spearman
corr_matrix = df_corr[vars_corr].corr(method='spearman')

# Visualización
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlación Spearman entre variables seleccionadas")
plt.tight_layout()
plt.show()
```
