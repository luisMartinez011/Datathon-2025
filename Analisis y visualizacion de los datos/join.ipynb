```python
import pandas as pd

transacciones = pd.read_csv("transacciones_limpias.csv")
clientes = pd.read_csv("clientes_limpios.csv")

# Unir los datasets por la columna "id"
clientes_transacciones = pd.merge(transacciones, clientes, on="id", how="left")  # Mantiene todos los clientes

# Ver los primeros registros para confirmar la unión
print(clientes_transacciones.head())
```
