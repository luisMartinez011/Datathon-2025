```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
transacciones = pd.read_csv("transacciones_modificado.csv")
transacciones.head()
```

```python
transacciones.info()
```

```python
transacciones = transacciones.rename(columns={"Columna1": "id"})
transacciones.head()
```

```python
transacciones["giro_comercio"] = transacciones["giro_comercio"].replace({"4121": "SERVICIOS DE TRANSPORTE POR AUTOMÓVIL"})
transacciones.giro_comercio.value_counts()
```

```python
transacciones = transacciones.dropna(subset=["giro_comercio"])
print(transacciones.info())
```

```python
transacciones.giro_comercio.value_counts()
```

```python
# Convertir espacios en blanco o cadenas vacías a NaN
transacciones["giro_comercio"] = transacciones["giro_comercio"].replace(r"^\s*$", np.nan, regex=True)
transacciones = transacciones.dropna(subset=["giro_comercio"])
print(transacciones.info())
```

```python
transacciones.giro_comercio.value_counts()
```

```python
transacciones["comercio"].value_counts()
```

```python
transacciones["comercio"] = transacciones["comercio"].replace({"7 ELEVEN": "7ELEVEN"})
transacciones["comercio"] = transacciones["comercio"].replace({"TOTAL PLAY": "TOTALPLAY"})
transacciones["comercio"] = transacciones["comercio"].replace({"APLAZ": "APLAZO"})
transacciones["comercio"] = transacciones["comercio"].replace({"DIDI FOOD": "DIDIFOOD"})
transacciones["comercio"] = transacciones["comercio"].replace({"MERCADO PAGO": "MERCADOPAGO"})
transacciones["comercio"] = transacciones["comercio"].replace({"SMART FIT": "SMARTFIT"})
transacciones["comercio"] = transacciones["comercio"].replace({"TOTAL PLAY": "TOTALPLAY"})
transacciones["comercio"] = transacciones["comercio"].replace({"WAL-MART": "WALMART"})
transacciones["comercio"].value_counts()
```

```python
transacciones["comercio"] = transacciones["comercio"].replace({"ATT": "AT&T", "MI ATT": "AT&T"})
transacciones["comercio"].value_counts()
```

```python
transacciones["giro_comercio"] = transacciones["giro_comercio"].str.replace(" ", "_")
transacciones["giro_comercio"].value_counts()
```

```python
transacciones["comercio"] = transacciones["comercio"].str.replace(" ", "_")
transacciones["comercio"].value_counts()
```

```python
transacciones.to_csv("transacciones_limpias.csv", index=False)
```
